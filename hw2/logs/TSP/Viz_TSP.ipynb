```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
## collecting best paths:
best_paths = []
```

```python
## Looking for the best MIMIC output
paths = glob.glob('*MIMIC*.csv')
print(paths)
```

```python
plt.figure()
for i,path in enumerate(paths):
    df = pd.read_csv(path)
    plt.plot(df['fitness'],label=i)

plt.title('Change in fitness function value in MIMIC')
plt.legend()
plt.show()
```

```python
best_paths.append(paths[0])
```

```python
plt.figure()
for i,path in enumerate(best_paths):
    df = pd.read_csv(path)
    plt.plot(df['fitness'],label=path.split('_')[1])
             
    best_obj = df.iloc[-1].fitness
    final_time = df.iloc[-1].time
    print(path.split('_')[1],best_obj,final_time)
    

plt.title('Max Obj Value Comparison - Traveling Salesman')
plt.legend()
plt.show()
```
